```python
print("Spark version:", spark.version)

current_catalog = spark.sql("SELECT current_catalog()").first()[0]
current_schema = spark.sql("SELECT current_schema()").first()[0]

print("Current catalog:", current_catalog)
print("Current schema:", current_schema)

spark.sql("SHOW CATALOGS").show(truncate=False)
```

```python
PROJECT_CATALOG = "workspace"
PROJECT_SCHEMA = "industrial_iot"

spark.sql(
    f"CREATE SCHEMA IF NOT EXISTS {PROJECT_CATALOG}.{PROJECT_SCHEMA}"
)

spark.sql(f"USE CATALOG {PROJECT_CATALOG}")
spark.sql(f"USE SCHEMA {PROJECT_SCHEMA}")

print(f"Project schema ready: {PROJECT_CATALOG}.{PROJECT_SCHEMA}")

spark.sql(
    "SELECT current_catalog() AS catalog, current_schema() AS schema"
).show()
```
